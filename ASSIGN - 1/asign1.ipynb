```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("startup_funding.csv")
data
```

```python
import pandas as pd

# Sample DataFrame
data_date = data["Date"]
df = pd.DataFrame(data)

# Convert to datetime with specified format
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y', errors='coerce')

# Extract the year
df['year'] = df['Date'].dt.year

# Display the result
print(df)
```

```python
data['Date'] = df['year']
```

```python
data_filled = data.fillna(0)
```

```python
data_filled
```

```python

data["Startup_Name"] = data["Startup_Name"].fillna(data['Startup_Name'].mode()) 
```

```python
data["City"] = data["City"].fillna(df['City'].mode()[0]) 
```

```python
data["City"].isnull().sum()
```

```python
data["Industry_Vertical"] = data["Industry_Vertical"].fillna(data['Industry_Vertical'].mode()[0]) 
data["Sub_Vertical"] = data["Sub_Vertical"].fillna(data['Sub_Vertical'].mode()[0])

data["Investors_Name"] = data["Investors_Name"].fillna(data["Investors_Name"].mode()[0])
data["Investment_Type"] = data["Investment_Type"].fillna(data['Investment_Type'].mode()[0])


# Handle missing values: Drop rows where critical data is missing (like 'Date' or 'Amount')
data = data.dropna(subset=['Date', 'Amount_USD'])

# Standardize column values (like Industry and Investment Type) for consistency
data['Industry_Vertical'] = data['Industry_Vertical'].str.lower().str.strip()
data['Investment_Type'] = data['Investment_Type'].str.lower().str.strip()

data["Remarks"] = data["Remarks"].fillna(data['Remarks'].mode()[0])
data["Date"] = data["Date"].fillna(data["Date"].mode()[0])

```

```python
data.head()
```

```python
data.dtypes
```

```python
data["Amount_USD"] = data['Amount_USD'].str.replace(',', '', regex=False)
```

```python
data['Amount_USD'] = pd.to_numeric(data['Amount_USD'], errors='coerce')
```

```python
data['Amount_USD'] = data['Amount_USD'].fillna(0).astype(int)
data["Date"] = data["Date"].astype(int)
# Step 3: Convert to int (will work now since non-numeric values are NaN)
```

```python
plt.figure(figsize=(12, 8))  # Adjust the width and height
plt.bar(data['Date'], data['Amount_USD'], color='skyblue')

# Adding titles and labels
plt.title('Funding Amount Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Amount in USD', fontsize=12)

# Customize ticks
plt.xticks(rotation=45, fontsize=10)  # Rotate x-axis labels for better readability
plt.yticks(fontsize=10)

# Add a grid
plt.grid()

# Display legend
plt.legend()

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

```python


# Histogram
plt.figure(figsize=(12, 8))
plt.hist(data['Amount_USD'], bins=10, color='blue', alpha=0.7, edgecolor='black')
plt.title('Histogram of Amounts', fontsize=16)
plt.xlabel('Amount in USD', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

```python
# Step 1: Sort the DataFrame by Amount_USD
top_startups = data.sort_values(by='Amount_USD', ascending=False).head(20)

# Step 2: Create the Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(top_startups['Amount_USD'], labels=top_startups['Startup_Name'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Amounts by Top 10 Startups', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Count the number of startups by industry
industry_counts = data["Industry_Vertical"].value_counts().head(10)  # Top 10 industries

# Plot a bar chart
plt.figure(figsize=(12, 6))
industry_counts.plot(kind='bar', color='skyblue')
plt.title('Top 10 Preferred Industries for Startups')
plt.xlabel('Industry')
plt.ylabel('Number of Startups')
plt.xticks(rotation=45)
plt.show()

```

```python
# Count the number of startups by city/location
location_counts = data['City'].value_counts().head(10)  # Top 10 cities

# Plot a bar chart
plt.figure(figsize=(12, 6))
location_counts.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Cities with Most Startups')
plt.xlabel('City')
plt.ylabel('Number of Startups')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count the top investors
investor_counts = data["Investors_Name"].value_counts().head(10)  # Top 10 investors

# Plot a bar chart
plt.figure(figsize=(12, 6))
investor_counts.plot(kind='bar', color='orange')
plt.title('Top 10 Investors in Indian Startups')
plt.xlabel('Investor')
plt.ylabel('Number of Investments')
plt.xticks(rotation=45)
plt.show()
```

```python
funding_type_counts = data["Investment_Type"].value_counts()

threshold = 14
filtered_funding_types = funding_type_counts[funding_type_counts >= threshold]

# Plot the filtered pie chart
plt.figure(figsize=(8, 8))
filtered_funding_types.plot(kind='pie', autopct='%1.1f%%', startangle=90)

# Add title and clean up the plot
plt.title('Distribution of Significant Funding Types')
plt.ylabel('')  # Remove y-axis label for a cleaner look

# Show the plot
plt.show()
```
